## Explicação da hipótese e mecanismo

O acordo do NAFTA, assinado em 1994, criou uma situação de livre comércio entre os três países da américa do Norte: Estados Unidos, México e Canadá. 

Para os Estados Unidos e o Canadá, sua atuação no bolco cria uma relação de dependência econômica com os outros países. Portanto, sua atuação no bloco se faz por aumentar suas exportações e aplicar o seu mercado em relção aos outros países do tratado. Além disso, é importante ressaltar o "boom" na economia devido às filiais americanas e canadenses instaladas no México, aproveitando da mão de obra barata. De acordo com Instituto Peterson de Economia Internacional (PIIE), os EUA incorporam uma riqueza anual no valor 
de US$ 127 bilhões todos os anos graças ao crescimento “extra” do comércio propiciado pelo NAFTA.

Apesar do México ser o menos desenvolvido dos três, ele ainda teve um benefício econômico, crescimento de empregos - devido às filiais americanas - e mercado para exportação.

Com tantos crescimentos econômicos vindo dos três países, os avanços políticos e, como consequência um crescimento do IDH, é possível e coerente concluír que haveria um aumento da taxa vegetativa devido ao tratado.

## Comparação entre o RNB e o IDH

Considerando o que foi dito na hipótese, é necessário provar que haja um aumento do IDH com o aumento da Renda Nacional Bruta. Caso essa informação seja verdadeira, apesar de não poder afirmar que o aumento do IDH se deve exclusivamente pelo RNB, há ainda uma relação econômica entre as duas.

```python
Y_idh = IDH_MUNDI.mean()
Y_gni = GNI_MUNDI.mean()

data1 = Y_idh
data2 = Y_gni
fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.set_xlabel('Anos')
ax1.set_ylabel('IDH mundial', color=color)
ax1.plot(data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax2 = ax1.twinx() 
color = 'tab:blue'
ax2.set_ylabel('GNI Mundial', color=color)  
ax2.plot( data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)
fig.tight_layout()  
plt.grid(True)
plt.show()


Y_idh_Nafta = IDH_NAFTA.mean()
Y_gni_Nafta = GNI_NAFTA.mean()

data1 = Y_idh_Nafta
data2 = Y_gni_Nafta
fig, ax1 = plt.subplots()
color = 'tab:green'
ax1.set_xlabel('Anos')
ax1.set_ylabel('IDH NAFTA', color="green")
ax1.plot(data1, color=color)
ax1.tick_params(axis='y', labelcolor="green")
ax2 = ax1.twinx() 
color = 'tab:purple'
ax2.set_ylabel('GNI NAFTA', color="purple")  
ax2.plot( data2, color=color)
ax2.tick_params(axis='y', labelcolor="purple")
fig.tight_layout()  
plt.grid(True)
plt.show()
```

Como pode-se observar, apesar de não ser correto afirmar que o aumento do IDH se deve exclusivamente ao crescimento do GNI, ainda há uma correlação.

## PAÍSES ANTES E DEPOIS DO NAFTA

```python
Mexico = GNI.loc["Mexico", [1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001]] 
USA = GNI.loc["United States", [1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001]]
Canada= GNI.loc["Canada",[1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001]]
Ym = Mexico
Xm = Ym.index
Yu = USA
Xu = Yu.index
Yc = Canada
Xc= Yc.index
plt.figure(figsize=(10,5))
plt.xlabel ('anos')
plt.ylabel ('RNB')
plt.plot(Xu,Yu, label = "USA")
plt.plot(Xm,Ym, label = "Mexico")
plt.plot(Xc,Yc, label = "Canada")
plt.legend(loc='upper left')
plt.show()
```

É possivel obeservar, entretanto, que não há muita diferença entre o GNI dos países antes e depois de 1994 (Quando o nafta foi formado) Por isso é coerente recorrer ao PIB, para observar como estes países evoluíram com o mercado exterior

```python
Mexico = PIB.loc["Mexico", [1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001]] 
USA = PIB.loc["United States", [1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001]]
Canada= PIB.loc["Canada",[1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001]]
Ym = Mexico
Xm = Ym.index
Yu = USA
Xu = Yu.index
Yc = Canada
Xc= Yc.index
plt.figure(figsize=(10,5))
plt.xlabel ('anos')
plt.ylabel ('PIB')
plt.plot(Xu,Yu, label = "USA")
plt.plot(Xm,Ym, label = "Mexico")
plt.plot(Xc,Yc, label = "Canada")
plt.legend(loc='upper left')
```

É possível observar que, apesar do acordo com o Estados Unidos e o Canadá, o PIB do México sofreu uma brusca decaída, porém nem um ano depois ela subiu ate ultrapassar o seu recorde em 1990. O que indica que, de fato, houve uma melhora na economía de todos os três países.
